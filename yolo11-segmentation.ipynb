```python
from ultralytics import YOLO

#Load pretrained model
model = YOLO("yolo11n-seg.pt")
```

```python
#Train the model om the carparts dataset
result = model.train(data = "carparts-seg.yaml",epochs = 10,imgsz=640)
```

```python
results = model.val()
```

```python
results = model.predict("/kaggle/working/datasets/carparts-seg/images/test/car42_jpg.rf.26816b314b704b50cd0270a11bc3c2ee.jpg")
```
